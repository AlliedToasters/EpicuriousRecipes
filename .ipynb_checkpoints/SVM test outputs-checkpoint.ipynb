```python
for category in categories:
    new_cat = []
    for ngram in categories[category]:
        for ft in bin_fts:
            if ngram in ft:
                new_cat.append(ft)
    categories[category] = list(set(new_cat))
    
remaining_ft = bin_fts.copy()
    
print(len(remaining_ft))

for cat in categories:
    for feature in categories[cat]:
        if feature in remaining_ft:
            remaining_ft.remove(feature)
            
```

```python
raw_data['ft_count'] = raw_data[bin_fts].sum(axis=1) #Total number of binary features that each recipe has

for cat in categories:
    raw_data[cat+'_count'] = raw_data[categories[cat]].sum(axis=1)

#fig = plt.figure(figsize=(10, 10))
#for cat in categories:
#    plt.hist(raw_data[cat+'_count'], alpha=.5, label=cat)
#plt.legend();
```

```python
raw_data['zero_rating'] = np.where(raw_data.rating == 0, 1, 0)
raw_data['4+_rating'] = np.where(raw_data.rating > 4, 1, 0)

print('ratings above zero: ', (len(raw_data[raw_data['zero_rating']==1])/len(raw_data)))
print('ratings above four: ', (len(raw_data[raw_data['4+_rating']==1])/len(raw_data)))
```

```python
categ = []
score = []
count = 0
for cat in categories:
    categ.append(cat)
    X = raw_data[[cat+'_count', 'ft_count']]
    Y = raw_data['zero_rating']
    mod = SVC(kernel='linear', class_weight='balanced')
    mod = mod.fit(X, Y)
    score.append(mod.score(X, Y))
    count += 1
    print(count, '/', len(categories), ' ', cat, ' complete: ', score[-1])
    

result = pd.DataFrame()
result['category'] = categ
result['score'] = score

print(result.head(len(result)))
```

```python
samp = raw_data.sample(frac=.5)

features = ['drinks']
others = list(categories.keys()) + ['ingredient']
for feat in others:
    if feat in features:
        others.remove(feat)
        
categ = []
score = []
count = 0
for cat in others:
    categ.append(cat)
    X = samp[features + [cat+'_count']]
    Y = samp['4+_rating']
    mod = SVC(class_weight='balanced')
    scores = pd.Series(cross_val_score(mod, X, Y, cv = 5))
    score.append(str(scores.mean())[:5] + ' +/-' + str(scores.std()))
    count += 1
    print(count, '/', len(others), ' ', cat, ' complete: ', score[-1])
    

result = pd.DataFrame()
result['category'] = categ
result['score'] = score

print(result.head(len(result)))
```
